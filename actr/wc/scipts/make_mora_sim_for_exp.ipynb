```python
import pandas as pd
import numpy as np
import itertools
```

```python
moras = "a,i,u,e,o,ja,ju,jo,ka,ki,ku,ke,ko,kja,kju,kjo,ga,gi,gu,ge,go,gja,gju,gjo,sa,si,su,se,so,sja,sju,sjo,za,zi,zu,ze,zo,zja,zju,zjo,ta,ti,tu,te,to,tja,tju,tjo,da,de,do,na,ni,nu,ne,no,nja,nju,njo,ha,hi,hu,he,ho,hja,hju,hjo,pa,pi,pu,pe,po,pja,pju,pjo,ba,bi,bu,be,bo,bja,bju,bjo,ma,mi,mu,me,mo,mja,mju,mjo,ra,ri,ru,re,ro,rja,rju,rjo,wa"#,N,R,Q"
moras = moras.split(",")

print(len(moras))

mora_combi = list(itertools.combinations(moras, 2))
print(len(mora_combi))
#print(mora_combi)

df_table = pd.read_csv("DF_table.csv", index_col=0)
df_table
```

```python
def cos_sim(v1, v2):
    return np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))
```

```python
#算出される類似度の最大値，最小値はこれとは異なるが，
#コサイン類似度の変域を0~-1に正規化するという意味でこれでいいはず
mora_csim_min = -1
#mora_csim_max = 0.9971715121772409
mora_csim_max = 1
max_sim = 0
min_sim = -1

#0~-1
def normarized_sim(v):
    return ((v - mora_csim_min) / (mora_csim_max - mora_csim_min)) * (max_sim - min_sim) + min_sim
```

# sim1 (Advanced Robotics Table 2)

```python
df_table1 = df_table.replace({"+": 2, "-": 1, np.nan: 0})
cons_table1 = df_table1.drop(["a","i","u","e","o"], axis=1)

mean_cons = cons_table1.mean(axis=1).values.tolist()
print(mean_cons)

df_table1
cons_table1
```

```python
def convert_mora2vec_sim1(v):
    result = []
    if len(v) == 1:
        result = df_table1[v].values.tolist() + mean_cons + [0] #isYoon = 0
    elif len(v) == 2:
        result = df_table1[v[0]].values.tolist() + df_table1[v[1]].values.tolist() + [0] #isYoon = 0
    elif len(v) == 3:
        result = df_table1[v[0]].values.tolist() + df_table1[v[2]].values.tolist() + [1] #isYoon = 1
    else:
        print("何かがおかしい！")
    return result

```

```python
sim_list = []
for conbi in mora_combi:
    tmp = []
    v1 = convert_mora2vec_sim1(conbi[0])
    v2 = convert_mora2vec_sim1(conbi[1])
    
    #print("calc:" + str(v1) + " and " + str(v2))

    sim = cos_sim(v1, v2)
    print(sim)

    tmp = [conbi[0], conbi[1], sim]
    sim_list.append(tmp)
#print(sim_list)
tmp = [x[2] for x in sim_list]
```

```python
with open("mora_sim1.lisp", mode="a") as output:
    output.write("(format t \"~A:mora similarities loading~%\" (current-model))\n")
    output.write("(defvar *mora-sim-list1*)\n")
    output.write("(setf *mora-sim-list1* '(\n")
    for i in sim_list:
        output.write("  (" + i[0] + " " + i[1] + " " + str(normarized_sim(i[2])) + ")\n")
    output.write("))\n")
```

# sim2 (Advanced Robotics Table 3)

```python
df_table2 = df_table.replace({"+": 1, "-": -1, np.nan: 0})

df_table2
```

```python
def convert_mora2vec_sim2(v):
    result = []
    if len(v) == 1:
        result = df_table2[v].values.tolist()
    elif len(v) == 2:
        result = np.mean([df_table2[v[0]].values.tolist(), df_table2[v[1]].values.tolist()], axis=0)
    elif len(v) == 3:
        result = np.mean([df_table2[v[0]].values.tolist(), df_table2[v[1]].values.tolist(), df_table2[v[2]].values.tolist()], axis=0)
    else:
        print("何かがおかしい！")
    return result
```

```python
sim_list = []
for conbi in mora_combi:
    tmp = []
    v1 = convert_mora2vec_sim2(conbi[0])
    v2 = convert_mora2vec_sim2(conbi[1])
    #print("v2: " + str(v2))

    sim = cos_sim(v1, v2)


    tmp = [conbi[0], conbi[1], sim]
    sim_list.append(tmp)
#print(sim_list)
```

```python
with open("mora_sim2.lisp", mode="a") as output:
    output.write("(format t \"~A:mora similarities loading~%\" (current-model))\n")
    output.write("(defvar *mora-sim-list2*)\n")
    output.write("(setf *mora-sim-list2* '(\n")
    for i in sim_list:
        output.write("  (" + i[0] + " " + i[1] + " " + str(normarized_sim(i[2])) + ")\n")
    output.write("))\n")
```
